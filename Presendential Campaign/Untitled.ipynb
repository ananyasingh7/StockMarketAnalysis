```python
import pandas as pd 
pd.set_option('display.max_columns', 999)
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
big_df=pd.read_csv('C:\\Users\\Meire\\Documents\\College\\4th year\\CS301\\Project\\big_five_stocks.csv')
```

```python
big_df['year'] = pd.DatetimeIndex(big_df['date']).year
big_df['month'] = pd.DatetimeIndex(big_df['date']).month
#big_5.tail
```

```python
Y15=big_df.loc[(big_df['year'] == 2015) | (big_df['year'] == 2016) | (big_df['year'] == 2017)]
```

```python
AMZN=Y15.loc[Y15['name'] == 'AMZN'].groupby('year')
AMZN15=AMZN.get_group(2015)
AMZN15M=AMZN15.groupby('month')[["close"]].mean().reset_index()
AMZN15M['year']=2015
AMZN15M
```

```python
AMZN=Y15.loc[Y15['name'] == 'AMZN'].groupby('year')
AMZN16=AMZN.get_group(2016)
AMZN16M=AMZN16.groupby('month')[["close"]].mean().reset_index()
AMZN16M['year']=2016
AMZN16M
```

```python
AMZN=Y15.loc[Y15['name'] == 'AMZN'].groupby('year')
AMZN17=AMZN.get_group(2017)
AMZN17M=AMZN17.groupby('month')[["close"]].mean().reset_index()
AMZN17M['year']=2017
AMZN17M
```

```python
AMZN_df = pd.concat([AMZN15M, AMZN16M, AMZN17M])
AMZN_df
```

```python
AAPL=Y15.loc[Y15['name'] == 'AAPL'].groupby('year')
AAPL16=AAPL.get_group(2016)
AAPL16=AAPL16.groupby('month')[["close"]].mean().reset_index()
AAPL16['year']=2016
AAPL16
```

```python
AAPL=Y15.loc[Y15['name'] == 'AAPL'].groupby('year')
AAPL17=AAPL.get_group(2017)
AAPL17=AAPL17.groupby('month')[["close"]].mean().reset_index()
AAPL17['year']=2017
AAPL17
```

```python
AAPL_df = pd.concat([AAPL15, AAPL16, AAPL17])
AAPL_df
```

```python
FB=Y15.loc[Y15['name'] == 'FB'].groupby('year')
FB15=FB.get_group(2015)
FB15=FB15.groupby('month')[["close"]].mean().reset_index()
FB15['year']=2015
FB15
```

```python
FB=Y15.loc[Y15['name'] == 'FB'].groupby('year')
FB16=FB.get_group(2016)
FB16=FB16.groupby('month')[["close"]].mean().reset_index()
FB16['year']=2016
FB16
```

```python
FB=Y15.loc[Y15['name'] == 'FB'].groupby('year')
FB17=FB.get_group(2017)
FB17=FB17.groupby('month')[["close"]].mean().reset_index()
FB17['year']=2017
FB17
```

```python
FB_df = pd.concat([FB15, FB16, FB17])
FB_df
```

```python
FB_df['date']=pd.to_datetime(FB_df['year'].astype(str) + FB_df['month'].astype(str), format='%Y%m')
AAPL_df['date']=pd.to_datetime(AAPL_df['year'].astype(str) + AAPL_df['month'].astype(str), format='%Y%m')
AMZN_df['date']=pd.to_datetime(AMZN_df['year'].astype(str) + AMZN_df['month'].astype(str), format='%Y%m')
```

```python
plt.plot(FB_df['date'].dt.strftime('%Y%m'), FB_df['close'], label='Facebook', color='blue')
#plt.plot(AMZN_df['date'].dt.strftime('%Y%m'), AMZN_df['close'], color='green')
plt.plot(AAPL_df['date'].dt.strftime('%Y%m'), AAPL_df['close'],label='Apple', color='red')
plt.gcf().autofmt_xdate()
plt.gcf().set_size_inches(14, 8)
plt.ylabel('Average close values ($)')
plt.xlabel('Date')
plt.title('Average Monthly Stock Close Value for Facebook and Apple during 2016,2017,2018')
plt.tight_layout()
plt.legend()
plt.show()
```
